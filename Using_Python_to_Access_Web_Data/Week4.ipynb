py3：unicode和str的type都是str；bytes string則會特別辨認
1. 要與network socket 溝通，會傳送bytes，所以我們必須encode py3的strings將它變成byte string<br>

##### 結論：
encode()：str -> byte <br>
decode()：byte -> str

### urllib

```python
import urllib.request, urllib.parse, urllib.error
file = urllib.request.urlopen('http://data.pr4e.org/romeo.txt')
counts = dict()
for line in file:
    words = line.decode().split() #byte str-> Unicode char str
    print(line)
    print(words)
    for word in words:
        counts[word] = counts.get(word,0)+1
print('\n',counts)
```

### Web scraping
A program pretends to be a browser and retrieves web pages<br>
- `BeautifulSoup` 物件：包含了整個 HTML 文件的結構樹，有了這個結構樹之後，就可以輕鬆找出資料了<br>
- 網址：https://www.crummy.com/software/BeautifulSoup/bs4/doc/<br>

```python
import urllib.request,urllib.parse,urllib.error
from bs4 import BeautifulSoup
url = input('Enter - ')
html = urllib.request.urlopen(url).read()
soup = BeautifulSoup(html,'html.parser')
print(soup.prettify())
```

```python
# 擷取所有anchor tags
tags = soup('a')
for tag in tags:
    print(tag.get('href',None))
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input('Enter - ')
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')

tags = soup('span')
for tag in tags:
    print('TAG:',tag)
    print('URL:',tag.get('href', None))
    print('Contents:',tag.contents[0])
    print('Attrs:',tag.attrs)
```

### Practice 1

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input('Enter - ')
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')

tags = soup('span')
count = 0
total = 0
for tag in tags:
    count=count+1
    num = int(tag.contents[0])
    total = total+num

print('Count',count)
print('Sum',total)
```

### Practice 2

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input('Enter URL: ')
count = int(input('Enter count: '))+1
pos = int(input('Enter position: '))

while(count):
    html = urllib.request.urlopen(url, context=ctx).read()
    soup = BeautifulSoup(html, 'html.parser')
    tags = soup('a')
    print('Retrieving: ',url)
    count = count-1
    url = tags[pos-1].get('href', None)
```
