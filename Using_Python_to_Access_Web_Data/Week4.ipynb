py3：unicode和str的type都是str；bytes string則會特別辨認
1. 要與network socket 溝通，會傳送bytes，所以我們必須encode py3的strings將它變成byte string<br>
2. 當我們要read data時，則要decode這些byte type字元，在py3用string(unicode/ASCII)呈現<br>
##### 結論：
encode()：str -> byte <br>
decode()：byte -> str

### urllib

```python
import urllib.request, urllib.parse, urllib.error
file = urllib.request.urlopen('http://data.pr4e.org/romeo.txt')
counts = dict()
for line in file:
    words = line.decode().split() #byte str-> Unicode char str
    print(line)
    print(words)
    for word in words:
        counts[word] = counts.get(word,0)+1
print('\n',counts)
```
